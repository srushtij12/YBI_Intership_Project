**Title Of Project **: Women Cloth Reviews Prediction With Multi Nomial Naive Bayes

Objective : Muti Nomial Bayes Classifier is Suitable for classificatin with descrete features. The multi Nomial Distribution normally requires integer feature counts. However in practice, Fractional counts such as tf-idf may also work.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Women Clothing E-Commerce Review.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review",inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

```python
df.columns
```

```python
x=df['Review']
```

```python
y=df['Rating']
```

```python
df['Rating'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,stratify=y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(lowercase=True,analyzer='word',ngram_range=(2,3),stop_words='english',max_features=5000)
```

```python
x_train =cv.fit_transform(x_train)
```

```python
cv.get_feature_names_out()
```

```python
x_train.toarray()
```

```python
x_test=cv.fit_transform(x_test)
```

```python
cv.get_feature_names_out()
```

```python
x_test.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
model.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
